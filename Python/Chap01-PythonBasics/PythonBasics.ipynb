### Task 3

Rock, Paper, Scissors Game 
Objective: Build a simple Rock-Paper-Scissors game where the player competes with the computer. 
This task will test your understanding of: 
● Input/output 
● Conditional logic 
● Random module 
● Loops 
● Exception handling 

Instructions: 
1. The user should enter their choice: rock, paper, or scissors.
2. The computer should randomly select one of these.
3. Compare both and print who wins.
4. Ask if the player wants to play again- if yes, repeat; if no, exit.
5. Handle invalid inputs gracefully using try-except or condition checks.

```python
import random

def get_computer_choice():
    return random.choice(["rock", "paper", "scissors"])

def decide_winner(user, computer):
    if user == computer:
        return "It's a tie!"
    elif (user == "rock" and computer == "scissors") or \
         (user == "paper" and computer == "rock") or \
         (user == "scissors" and computer == "paper"):
        return "You win!"
    else:
        return "Computer wins!"

while True:
    try:
        user_choice = input("\nEnter rock, paper, or scissors: ").lower()

        if user_choice not in ["rock", "paper", "scissors"]:
            raise ValueError("Invalid choice")

        computer_choice = get_computer_choice()

        print("Computer chose:", computer_choice)

        result = decide_winner(user_choice, computer_choice)
        print(result)

    except ValueError as e:
        print("Error:", e)
        continue

    # Ask to play again
    again = input("Play again? (yes/no): ").lower()
    if again != "yes":
        print("Thanks for playing!")
        break
```

### Task 4 Word guessing game

Word Guessing Game (Mini Hangman) 
Objective: Create a word-guessing game where the player tries to guess the letters of a hidden word. 

This will test your knowledge on the following topics : 

● Loops 
● Conditional logic 
● String manipulation 
● Exception handling 
● Logical thinking 

Instructions: 
1. Create a list of random words (like “python”, “machine”, “learning”, “bridge”).
2. Randomly select one word for the game.
3. Show the word as blanks (_ _ _ _) and let the player guess one letter at a time.
4. Keep track of wrong guesses (max 6 attempts).
5. End the game when the player either guesses the full word or uses all chances.

```python
import random

# Step 1: Word list
words = ["python", "machine", "learning", "bridge"]

# Step 2: Random word selection
secret_word = random.choice(words)

guessed_letters = []
wrong_attempts = 0
max_attempts = 6

print("Welcome to Word Guessing Game!")
print("Guess the word letter by letter.")

# Step 3: Game loop
while wrong_attempts < max_attempts:
    
    # Display current progress
    display_word = ""
    for letter in secret_word:
        if letter in guessed_letters:
            display_word += letter + " "
        else:
            display_word += "_ "
    
    print("Word:", display_word.strip())
    print("Wrong attempts left:", max_attempts - wrong_attempts)
    
    # Check win condition
    if "_" not in display_word:
        print("Congratulations! You guessed the word:", secret_word)
        break

    try:
        guess = input("Enter a letter: ").lower()

        # Input validation
        if len(guess) != 1 or not guess.isalpha():
            raise ValueError("Enter only ONE alphabet letter.")

        if guess in guessed_letters:
            print("You already guessed that letter.")
            continue

        guessed_letters.append(guess)

        # Step 4: Check guess
        if guess not in secret_word:
            wrong_attempts += 1
            print("Wrong guess!")

    except ValueError as e:
        print("Error:", e)

# Step 5: Lose condition
if wrong_attempts == max_attempts:
    print("Sorry! The game is over! The word was:", secret_word)
```

```python
import random
import threading
import time

topics = {
    "fruits": ["apple","banana","mango","orange","grapes","pineapple","watermelon","papaya","kiwi","peach",
               "cherry","pear","plum","apricot","guava","fig","pomegranate","blueberry","raspberry","lemon"],

    "vegetables": ["carrot","potato","tomato","onion","cabbage","spinach","broccoli","cauliflower","pepper","radish",
                   "turnip","pumpkin","cucumber","lettuce","peas","beans","garlic","ginger","zucchini","beetroot"],

        "countries": [
        "india","china","unitedstates","indonesia","pakistan","brazil","nigeria","bangladesh","russia","mexico",
        "japan","ethiopia","philippines","egypt","vietnam","congo","turkey","iran","germany","thailand",
        "unitedkingdom","france","italy","tanzania","southafrica","myanmar","kenya","southkorea","colombia","spain",
        "argentina","uganda","ukraine","algeria","sudan","iraq","afghanistan","poland","canada","morocco",
        "saudiarabia","uzbekistan","peru","angola","malaysia","mozambique","ghana","yemen","nepal","venezuela",
        "madagascar","cameroon","australia","northkorea","ivorycoast","taiwan","niger","srilanka","burkinafaso","mali",
        "romania","malawi","chile","kazakhstan","zambia","guatemala","ecuador","syria","netherlands","senegal",
        "cambodia","chad","somalia","zimbabwe","guinea","rwanda","benin","burundi","tunisia","haiti",
        "belgium","jordan","greece","czechrepublic","portugal","sweden","azerbaijan","hungary","belarus","uae",
        "austria","switzerland","israel","denmark","finland","norway","ireland","newzealand","singapore","qatar"
    ],

    "colours": [
        "red","blue","green","yellow","orange","purple","pink","brown","black","white",
        "grey","violet","indigo","turquoise","cyan","magenta","maroon","beige","gold","silver"
    ],

    "households": [
        "table","chair","sofa","bed","cupboard","mirror","fan","light","curtain","pillow",
        "blanket","carpet","bucket","broom","mop","stove","fridge","microwave","washingmachine","television"
    ],

    "animals": [
        "lion","tiger","elephant","giraffe","zebra","monkey","bear","wolf","fox","kangaroo",
        "panda","leopard","cheetah","horse","cow","goat","sheep","dog","cat","rabbit"
    ],

    "birds": [
        "sparrow","pigeon","crow","eagle","parrot","peacock","owl","flamingo","duck","swan",
        "woodpecker","kingfisher","penguin","ostrich","falcon","hawk","canary","robin","seagull","heron"
    ],

    "sports": ["cricket","football","basketball","tennis","badminton","hockey","volleyball","baseball","rugby","golf",
               "swimming","boxing","cycling","skating","archery","karate","wrestling","surfing","skiing","rowing"],

    "languages": ["english","spanish","french","german","italian","hindi","chinese","japanese","korean","arabic",
                  "russian","portuguese","dutch","swedish","greek","turkish","thai","vietnamese","polish","hebrew"],

    "beverages": ["coffee","tea","milk","juice","lemonade","smoothie","soda","cola","espresso","latte",
                  "cappuccino","mocha","milkshake","cocktail","mocktail","water","icedtea","greentea","hotchocolate","buttermilk"]
}

category_hints = {

    "countries": {
        "india": "Asia", "brazil": "South America", "canada": "North America",
        "france": "Europe", "egypt": "Africa", "australia": "Oceania",
        "japan": "Asia", "germany": "Europe", "argentina": "South America"
    },

    "animals": {
        "lion": "Wild animal", "dog": "Domestic animal",
        "elephant": "Herbivore mammal", "tiger": "Carnivore mammal",
        "cow": "Domestic herbivore", "fox": "Wild omnivore"
    },

    "birds": {
        "eagle": "Flying bird", "penguin": "Non-flying bird",
        "duck": "Water bird", "owl": "Nocturnal bird",
        "sparrow": "Small flying bird"
    },

    "colours": {
        "red": "Warm colour", "blue": "Cool colour",
        "green": "Cool colour", "yellow": "Warm colour",
        "black": "Neutral colour", "white": "Neutral colour"
    },

    "households": {
        "bed": "Bedroom item", "sofa": "Living room item",
        "fridge": "Kitchen appliance", "mirror": "Bathroom item",
        "table": "Furniture item"
    }
}

def get_category_hint(topic, word):
    if topic in category_hints and word in category_hints[topic]:
        return category_hints[topic][word]
    return "No additional hint available"
    
topic_list = list(topics.keys())

def display_topics():
    print("\nAvailable Topics:")
    for i, topic in enumerate(topic_list, start=1):
        print(f"{i}. {topic.capitalize()}")

def get_topic():
    while True:
        try:
            choice = int(input("Choose topic number: "))
            if 1 <= choice <= len(topic_list):
                return topic_list[choice - 1]
            else:
                print("Invalid number!")
        except ValueError:
            print("Enter a valid number!")

def get_difficulty():
    while True:
        choice = input("Choose difficulty (easy/medium/hard): ").strip().lower()
        if choice in ["easy", "medium", "hard"]:
            return choice
        else:
            print("Invalid choice!")

def generate_hint(word, difficulty):
    length = len(word)

    if difficulty == "easy":
        if length <= 5:
            hide = 1
        elif length < 10:
            hide = 2
        elif length <= 15:
            hide = 3
        else:
            hide = 4

    elif difficulty == "medium":
        hide = length // 2

    else:  # hard
        hide = length - 2 if length > 2 else 1

    hide = min(hide, length)
    positions_to_hide = random.sample(range(length), hide)

    hint = ""
    for i in range(length):
        if i in positions_to_hide:
            hint += "_ "
        else:
            hint += word[i] + " "

    return hint.strip()

time_up = False

def timer(stop_event):
    time.sleep(60)
    if not stop_event.is_set():
        print("\nTIME'S UP! Game Over!")
        stop_event.set()

def play_game(word, difficulty, topic):

    max_attempts = 5
    attempts_used = 0

    hint = generate_hint(word, difficulty)
    extra_hint = get_category_hint(topic, word)

    print("Hint : ", hint)
    print("Word Length:", len(word))
    print("Category Hint:", extra_hint)
    print("You have only 60 seconds!")

    # ✅ Create a fresh event for each game
    stop_event = threading.Event()

    # Start timer thread
    t = threading.Thread(target=timer, args=(stop_event,))
    t.daemon = True
    t.start()

    attempts_left = max_attempts

    while attempts_left > 0:

        if stop_event.is_set():
            print("The word was:", word)
            return

        guess = input(f"Guess the {topic}: ").strip().lower()
        attempts_used += 1

        if stop_event.is_set():
            print("TIME'S UP! Stopping game...")
            print("The word was:", word)
            return

        if guess == word:
            print(f"Fantastic! You guessed it in {attempts_used} attempt(s)!")
            stop_event.set()  # stop timer
            return
        else:
            attempts_left -= 1
            print(f"Wrong! Attempts left: {attempts_left}")

    print(f"Sorry! You lost the game! The word was: {word}")
    stop_event.set()  # stop timer

def main():
    while True:
        display_topics()
        topic = get_topic()
        difficulty = get_difficulty()

        word = random.choice(topics[topic])
        play_game(word, difficulty, topic)

        # Ask to continue
        choice = input("Do you want to continue? [Y/N]: ").strip().lower()
        if choice != 'y':
            print("Thanks for playing!")
            break


main()
```
