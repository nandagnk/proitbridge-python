## Sets
used to perform set operations like union, intersection, union all, minus 

1. one dimentional
2. Hetrogeneous
3. won't allow duplicate values
4. individual items are not accessble using index.
5. sort the items by default

```python
empty_set = set()
empty_set
```

```python
type(empty_set)
```

```python
multiples_of_two = {2,4,6,8,10,12,2}
multiples_of_two
```

```python
multiples_of_three = {3,6,9,12,15,18}
multiples_of_three
```

```python
# union is similar to A union B in set operations
multiples_of_two.union(multiples_of_three)
```

```python
# intersection is similar to A intersect B in set operations
multiples_of_two.intersection(multiples_of_three)
```

```python
# add method add an element to the set
multiples_of_two.add(14)
multiples_of_two
```

```python
multiples_of_two.add(18)
multiples_of_two
```

```python

multiples_of_two.add(16)
multiples_of_two
```

```python
# discard removes the element from the set 
multiples_of_two.discard(12)
multiples_of_two
```

```python
# difference is equivalent to A-B in set operations
multiples_of_two.difference(multiples_of_three)
```
