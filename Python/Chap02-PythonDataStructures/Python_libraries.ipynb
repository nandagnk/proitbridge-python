```python
marks_df = pd.DataFrame(data = np.repeat(a = marks_data, repeats=1, axis=0))
labels = marks_df[1]  # or combine names if you want
plt.bar(labels, marks_df[3])

plt.xlabel("Students")
plt.ylabel("Maths Marks")
plt.title("Maths Marks Comparison")

plt.show()
```

```python
student = marks_df[marks_df[0] == 1001].iloc[0]
```

```python
student
```

```python
subjects = [3, 4, 5, 6, 7]
marks = student[subjects]
marks
```

```python
import matplotlib.pyplot as plt

plt.bar(subjects, marks)

plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title(f"Marks of {student[1]}")

plt.show()
```
